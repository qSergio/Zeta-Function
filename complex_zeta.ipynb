```python
import mpmath
import numpy
import colorsys
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 2-d array of points
real_points = 700
imag_points = 700
points = numpy.zeros((real_points, imag_points,3), dtype='float64')

# viewport on complex plane
rmin = 1.01
rmax = 6.01
imin = -2.5
imax = 2.5
```

```python
# calculate zeta(s)

for rp in range(real_points):
    for ip in range(imag_points):
        real = rmin + ((rmax - rmin) * rp/real_points)
        imag = imin + ((imax - imin) * ip/imag_points)
        z = mpmath.zeta(complex(real,imag))
        rgb = colorsys.hsv_to_rgb(numpy.abs(z), 0.8, 0.8)
        points[ip,rp][0:3] = rgb
        pass
    pass
```

```python
# normalise array to [0..1]
points += - points.min()
points /= points.max()
print(points.min(), points.max())
```

```python
# plot graph, you can right-click and save as an image
plt.figure(figsize = (15,15))
plt.imshow(points, aspect='equal', interpolation='bilinear', origin='lower', extent=[rmin, rmax, imin, imax])
```
